### Correção da Importação da Base

```python
# Ler arquivo sem rótulos de colunas
df = pd.read_csv('df_yields.txt', header = None)

# Definir rótulos para colunas
df = df.set_axis(['drop', 'date', 'M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', 'M9', 'M10'], axis = 1)

# Drop na primeira coluna
df = df.drop(['drop'], axis = 1)

# Transformar primeira coluna em index
df = df.set_index('date')
```

```python
df.tail()
```

```python
# Criar objeto excel writer
writer = pd.ExcelWriter('df_RNA.xlsx')

# Alocar df ao Excel
df.to_excel(writer)

# Salver o Excel
writer.save()
print('Dataframe salvo com sucesso!')
```
